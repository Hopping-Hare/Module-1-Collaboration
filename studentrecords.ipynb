```python
first_name = input("What is your first name? ")
last_name = input("What is your last name? ")
gpa = float(input("What is your GPA? "))

while last_name != "ZZZ":    
    if gpa >= 3.5:
        print("Student has made the Dean's List!")
        break
    if gpa >= 3.25:
        print("Student has made the Honor Roll!")
        break
```
